```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
G = nx.DiGraph()

G.add_edge('x1', 'x2', color='blue')
G.add_edge('x2', 'x1', color='red')
G.add_edge('x1', 'x3', color='blue')
G.add_edge('x2', 'x3', color='red')
G.add_edge('x3', 'x2', color='blue')
G.add_edge('x3', 'x3', color='black')

pos = nx.spring_layout(G) 
edges = G.edges()
edge_colors = [G[u][v]['color'] for u, v in edges]
edge_labels = nx.get_edge_attributes(G, 'label')

nx.draw_networkx_nodes(G, pos, node_size=700, node_color='skyblue', alpha=0.8)
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black', font_weight='bold')
nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color=edge_colors, arrowsize=20, connectionstyle='arc3,rad=0.1')

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Graphe Orienté avec Arêtes Distinctes")
plt.show()
```

```python
matrice = nx.to_numpy_array(G)
print("matrice d'adjacence")
print(matrice)
        
```

```python
M = np.linalg.matrix_power(matrice,4)
print(M)
```

```python
nombre_circuit_longueur_4 = (np.trace(M))
print(f" Le nombre de circuit de longueur 4 est {nombre_circuit_longueur_4}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_edge('X', 'A', color='black', weight=1)
G.add_edge('X', 'C', color='black', weight=1)
G.add_edge('C', 'D', color='black', weight=1)
G.add_edge('C', 'G', color='black', weight=1)
G.add_edge('D', 'E', color='black', weight=1)
G.add_edge('E', 'F', color='black', weight=1)
G.add_edge('A', 'G', color='black', weight=1)
G.add_edge('A', 'H', color='black', weight=1)
G.add_edge('A', 'B', color='black', weight=1)
G.add_edge('G', 'H', color='black', weight=1)
G.add_edge('B', 'H', color='black', weight=1)
G.add_edge('H', 'E', color='black', weight=1)
G.add_edge('H', 'F', color='black', weight=1)
G.add_edge('F', 'Y', color='black', weight=1)
G.add_edge('H', 'Y', color='black', weight=1)
G.add_edge('Y', 'B', color='black', weight=1)

pos = nx.spring_layout(G) 
edges = G.edges()
edge_colors = [G[u][v]['color'] for u, v in edges]

nx.draw_networkx_nodes(G, pos, node_size=700, node_color='skyblue', alpha=0.8)
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black', font_weight='bold')
nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color=edge_colors, arrowsize=20, connectionstyle='arc3,rad=0.1')

etiquettes_arretes = {(u, v): f"{G[u][v]['weight']}" for u, v in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=etiquettes_arretes)

plt.title("Graphe Orienté avec Couleurs d'Arêtes Distinctes")
plt.show()
```

```python
matrice = nx.to_numpy_array(G)
print("matrice d'adjacence")
print(matrice)
```

```python
distance_matrix = nx.floyd_warshall_numpy(G)
print("matrice de distance")
print(distance_matrix)
```

```python
chemin_plus_court = nx.dijkstra_path(G, 'X', 'Y')
longueur_chemin_plus_court = nx.dijkstra_path_length(G, 'X', 'Y')

print("Chemin le plus court:", chemin_plus_court)
print("Longueur du chemin le plus court:", longueur_chemin_plus_court)
```

```python
def nombre_chromatique(graphe):
    return nx.coloring.greedy_color(graphe, strategy="largest_first")
```

```python
coloration = nombre_chromatique(G)
nombre_chromatique = max(coloration.values()) + 1
print("Nombre chromatique du graphe:", nombre_chromatique)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

def greedy_adjacent_edge_coloring(graph):
    colors = {}
    
    for edge in graph.edges():
        neighbor_colors = set(colors.get(neighbor, None) for neighbor in graph.edges(edge[0]))
        
        color = 0
        while color in neighbor_colors:
            color += 1
        
        colors[edge] = color
    return colors

edge_colors = greedy_adjacent_edge_coloring(G)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=edge_colors.keys(), edge_color=list(edge_colors.values()), edge_cmap=plt.cm.rainbow)

plt.show()
```
